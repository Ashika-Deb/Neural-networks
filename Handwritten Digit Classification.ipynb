```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train
```

- The data is in the form of pixel values and not images per se
- Every image is stored as a numpy array

```python
X_test.shape
```

```python
y_train
```

```python
# can convert these pixel values back to the corresponding images as well
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

- Need to scale every pixel value in the range [0, 1] so that the NN weights converge faster

```python
model = Sequential()

model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

- Always use softmax in the output layer for MCC

```python
model.summary()
```

- No parameters in the flatten layer because no training occurs at the flatten layer

```python
model.compile(loss='sparse_categorical_crossentropy',
             optimizer='SGD')
```

- Labels are not one-hot encoded, thus, the chosen loss function

```python
model.fit(X_train, y_train, validation_split=0.2, epochs=100)
```

```python
y_prob = model.predict(X_test)
```

- Need to find the index position for which the probability value is maximum

```python
y_prob[0]
```

```python
import numpy as np
```

```python
np.argmax(y_prob[0])
```

```python
y_hat = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_hat)
```

- ML models which undergo hyper parameter tuning to a great extent gave a maximum accuracy of 0.96
- Without any tweaking, we are getting almost 98% accuracy

### Second model

```python
model2 = Sequential()

model2.add(Flatten(input_shape=(28, 28)))
model2.add(Dense(128, activation='relu'))
model2.add(Dense(32, activation='relu'))
model2.add(Dense(10, activation='softmax'))
```

```python
model2.summary()
```

```python
model2.compile(loss='sparse_categorical_crossentropy',
              optimizer='SGD',
              metrics=['accuracy'])
```

```python
history = model2.fit(X_train, y_train, epochs=25, validation_split=0.2)
```

```python
y_prob2 = model2.predict(X_test)
```

```python
y_hat2 = y_prob2.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_hat2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

- Starting to overfit as evident from the loss and accuracy curves
